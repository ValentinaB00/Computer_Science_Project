```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
df = pd.read_csv("dogs.csv")
df
```

```python
# (.at) selects a single scalar value in the DataFrame by label only
from numpy import NaN 
for i in range(0,len(df)):
  if len(df.loc[i]["stateQ"]) == 10:
    df.at[i,'accessed']=df.loc[i]["stateQ"]
    df.at[i,"stateQ"]=NaN
  if len(df.loc[i]["contact_country"]) > 2:
    df.at[i,'stateQ']=df.loc[i]["contact_country"]
    df.at[i,"contact_country"]=NaN
  if len(df.loc[i]["contact_state"]) > 2:
    df.at[i,'contact_zip']=df.loc[i]["contact_state"]
    df.at[i,"contact_state"]=NaN
  if len(df.loc[i]["contact_city"]) == 2:
    df.at[i,'contact_state']=df.loc[i]["contact_city"]
    df.at[i,"contact_city"]=NaN
  if len(df.loc[i]["posted"]) < 24:
    df.at[i,'contact_city']=df.loc[i]["posted"]
    df.at[i,"posted"]=NaN
  if len(df.loc[i]["status"]) == 24:
    df.at[i,'posted']=df.loc[i]["status"]
    df.at[i,"status"]=NaN
```

8. Fix the travels table so that the correct state is computed from the manual and the found fields. If manual is not missing, then it overrides what is stored in found.


```python
for i in range(0,len(dfTravel["manual"])):
  if str(dfTravel["manual"][i])=="nan":
    dfTravel.loc[i, "Correct"] = dfTravel["found"][i]
  else:
    dfTravel.loc[i, "Correct"] = dfTravel["manual"][i]
```

10. For each dog, compute the number of days from the posted day to the day of last access.


```python
df['posted']=pd.to_datetime(df['posted'],format='%Y-%m-%d %H:%M:%S',errors="coerce")
df['accessed']=pd.to_datetime(df['accessed'],format='%Y-%m-%d',errors="coerce")
```

```python
df['posted_day'] = df['posted'].dt.strftime('%Y-%m-%d')#non considero l'ora
df['posted_day']=pd.to_datetime(df['posted_day'],format='%Y-%m-%d')#Imposto come la colonna accessed
```

```python
df['number_of_day']=df['accessed']-df['posted_day']
df['number_of_day']
```

11. Partition the dogs according to the number of weeks from the posted day to the day of last access.


```python
df['number_of_week'] = 0
for i in range(0,len(df)):
  df.loc[i, 'number_of_week'] = int(df['number_of_day'][i].total_seconds()/86400/7)
```

```python
#group by per numero di settimane
pp=df.groupby("number_of_week")["id"]
```

```python
for a, b in pp:
  print(f"ID dog for {a} weeks")
  print("------------------------")
  print(b, end="\n\n")
```

12. Find for duplicates in the dogs dataset. Two records are duplicates if they have (1) same breeds and sex, and (2) they share at least 90% of the words in the description field. Extra points if you find and implement a more refined for determining if two rows are duplicates.

```python
!pip install fuzzywuzzy
from fuzzywuzzy import fuzz
```

```python
#df2 = pd.DataFrame(columns=df.columns)
```

```python
#df2.to_csv('duplicati_finale.csv', index=False)
```

```python
df2 = pd.DataFrame(columns=df.columns)
for i in range(0,len(df)):
  for j in range(0,len(df)):
    if i!=j:
      if df["breed_primary"][i]==df["breed_primary"][j]:
        if df["breed_secondary"][i]==df["breed_secondary"][j]:
          if df["sex"][i]==df["sex"][j]:
            if df["description"][j] is not NaN and df["description"][i] is not NaN: #Se non sono null vado a capire la similarità
              if fuzz.ratio(df["description"][i],df["description"][j]) > 90: #Similarità delle parole
                df2.loc[len(df2)] = df.iloc[j].tolist()
            elif df["description"][j] is NaN and df["description"][i] is NaN: #Se sono null entrambi vuol dire che sono uguali
              df2.loc[len(df2)] = df.iloc[j].tolist()

# Elimina le righe duplicate dal DataFrame originale
df2.drop_duplicates(subset='id', inplace=True)
```
